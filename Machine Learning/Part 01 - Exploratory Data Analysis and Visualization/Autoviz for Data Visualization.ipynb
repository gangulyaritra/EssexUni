```python
from autoviz.AutoViz_Class import AutoViz_Class
AV = AutoViz_Class()
```

```python
# Import Dataset.
import pandas as pd
dataset = pd.read_csv("CE802_Ass_2019_Data.csv")
dataset.head()
```

```python
filename = "CE802_Ass_2019_Data.csv"
sep = ','
target = "Class"

dft = AV.AutoViz(
    filename,
    sep,
    target,
    dataset,
    header=0,
    verbose=0,
    lowess=False,
    chart_format='svg',
    max_rows_analyzed=150000,
    max_cols_analyzed=30
)
```
