```
import matplotlib.pyplot as plt
import numpy as np

MCTS = [19, 21, 18, 24, 16, 15, 18, 35, 21, 21]
MCTS_DT = [1, 1, 2, 2, 0, 3, 0, 2, 0, 0]
DRAW = [80, 78, 80, 74, 84, 82, 82, 63, 79, 79]

n_groups = 10
# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.28
opacity = 0.8

rects1 = plt.bar(index, MCTS, bar_width, alpha = opacity, color = 'b', label = 'MCTS')
rects2 = plt.bar(index + bar_width, MCTS_DT, bar_width, alpha = opacity, color = 'r', label = 'MCTS + DT')
rects3 = plt.bar(index + bar_width + bar_width, DRAW, bar_width, alpha = opacity, color = 'g', label = 'DRAW')

plt.title('Agent Performance')
plt.xticks(index + bar_width, ('Iteration 1', 'Iteration 2', 'Iteration 3', 'Iteration 4', 'Iteration 5', 'Iteration 6', 'Iteration 7', 'Iteration 8', 'Iteration 9', 'Iteration 10'), rotation = 45)
plt.legend()
```
